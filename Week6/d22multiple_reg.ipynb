```python
# take any 3 suitable features and build the regression model
```

```python
x = df[['disp', 'hp', 'wt']]
y = df['mpg']
```

```python
x;
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x, y)
```

```python
lr.score(x, y)
```

```python
# feature selection
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
fs = SequentialFeatureSelector(LinearRegression(),
                              direction = 'backward',
                              )
```

```python
x = df.drop('mpg', axis = 1)
```

```python
fs.fit(x, y)
```

```python
fs.get_support()
```

```python
x = x.iloc[:, fs.get_support()]
```

```python
x
```

```python
lr.fit(x, y)
```

```python
lr.score(x, y)
```
