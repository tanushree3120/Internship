### Practice questions

##### 1. Import matplobtlin.pyplot and create a basic line plot showing fare values of the first 10 passengers.

```python
import matplotlib.pyplot as plt

first_10 = df.head(10)

plt.plot(first_10['PassengerId'], first_10['Fare'])  # Line plot
plt.title("Line Plot")
plt.xlabel("PassengerId")
plt.ylabel("Fare")
```

#### 2.Create a bar chart showing the number of passengers in each passengers class(pclass)

```python
df['Pclass'].value_counts().sort_index().plot(kind='bar')
plt.title("Passenger Count by Class")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.show()
```

### # 3.Histogram of 'Age' column

```python
df['Age'].plot(kind='hist', bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()
```

### 4.Scatter plot comparing 'Age' and 'Fare'

```python
plt.scatter(df['Age'], df['Fare'], color='green')
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

### 5.Pie chart: percentage of survivors vs non-survivors

### 7.Two subplots side-by-side

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('titanic')

fig, axs = plt.subplots(1, 2, figsize=(12, 5))


df.groupby('sex')['survived'].sum().plot(kind='bar', ax=axs[0], color=['skyblue', 'pink'])
axs[0].set_title('Survival Count by Gender')
axs[0].grid(True)


df['embarked'].value_counts().plot(kind='pie', ax=axs[1], autopct='%1.1f%%')
axs[1].set_title('Embarked Distribution')
axs[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

### 8.2x2 Subplots with 4 Titanic Visualizations

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line Plot
df.groupby('age')['fare'].mean().plot(ax=axs[0, 0])
axs[0, 0].set_title('Average Fare by Age')

# Bar Plot
df['class'].value_counts().plot(kind='bar', ax=axs[0, 1])
axs[0, 1].set_title('Passenger Class Count')

# Scatter Plot
axs[1, 0].scatter(df['age'], df['fare'], alpha=0.5, color='green')
axs[1, 0].set_title('Age vs Fare')
axs[1, 0].set_xlabel('Age')
axs[1, 0].set_ylabel('Fare')

# Histogram
df['age'].plot(kind='hist', bins=20, ax=axs[1, 1], color='orange')
axs[1, 1].set_title('Age Distribution')

plt.tight_layout()
plt.show()
```

### 9.Stacked Bar Plot (Survivors by Class)

```python
survived = df[df['survived'] == 1]['class'].value_counts()
not_survived = df[df['survived'] == 0]['class'].value_counts()

df_stacked = pd.DataFrame({'Survived': survived, 'Not Survived': not_survived})
df_stacked.plot(kind='bar', stacked=True, color=['green', 'red'])
plt.title('Survival by Passenger Class')
plt.ylabel('Count')
plt.show()
```

### 12. Custom Marker Styles in Scatter Plot

```python
plt.scatter(df['age'], df['fare'], 
            c='blue', marker='x', s=50, alpha=0.6)
plt.title('Age vs Fare (Custom Markers)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```
