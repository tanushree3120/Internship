```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
# histogram
```

```python
sns.histplot(tips['total_bill'], kde = False, bins = 15)
plt.title('total bill distribution');
```

```python
# kde plot
```

```python
sns.kdeplot(tips['total_bill'], fill = True)
plt.title("KDE of total bill");
```

```python
sns.histplot(tips['total_bill'], kde = True, bins = 15)
plt.title('total bill distribution');
```

```python
# boxplot for outlier detection
```

```python
sns.boxplot(tips['tip']);
```

```python
q3 = np.percentile(tips['tip'], 75)
q1 = np.percentile(tips['tip'], 25)
q3, q1
```

```python
iqr = q3 - q1
iqr
```

```python
ub = q3 + iqr * 1.5
lb = q1 - 1.5 * iqr
ub, lb
```

```python
# remove the outerliers
x = tips[(tips['tip'] > lb) & (tips['tip'] < ub)]
```

```python
sns.boxplot(x['tip']);
```
