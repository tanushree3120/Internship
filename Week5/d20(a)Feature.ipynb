```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df
```

```python
df.describe()
```

```python
# plot the scatter plot between age and estimated salary
```

```python
plt.title("distribution")
plt.xlabel("estimated salary")
plt.ylabel("age")
plt.scatter(df['EstimatedSalary'],df['Age'], c= df['Purchased']);
```

#### Normalization

```python
# separate the data
```

```python
x = df[['Age','EstimatedSalary']]
```

```python
x
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
x_scaled[:,0].max(), x_scaled[:, 0].min()
```

```python
x_scaled[:,1].max(), x_scaled[:,1].min()
```

```python
plt.title("distribution")
plt.xlabel("estimated salary")
plt.ylabel("age")
plt.scatter(x_scaled[:, 1], x_scaled[:, 0], c= df['Purchased']);
```
